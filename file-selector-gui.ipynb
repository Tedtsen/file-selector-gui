```python
import tkinter as tk
from tkinter import filedialog, Text
import os

root = tk.Tk()
root.minsize(400, 400)
# Add accepted file formats here with a 2-tuple entry
accepted_file_formats = [("text file", "*.txt"),
                         ("executables","*.exe"),
                         ("all files","*.*")]
selected_filepaths = []

#for i in range(20):
#    selected_filepaths.append("haha  some file.txt")

def onFrameConfigure(canvas):
    # Reset the scroll region to encompass the inner frame
    canvas.configure(scrollregion=canvas.bbox("all"))
    
def clear_list():
    for widget in list_frame.winfo_children():
        if isinstance(widget, tk.Frame):
            widget.destroy()

def build_list():
    for index, filepath in enumerate(selected_filepaths):
        line_frame = tk.Frame(list_frame, bg="gray")
        line_frame.grid(sticky="w")
        label = tk.Label(line_frame, text=filepath, bg="gray", padx=5, pady=5)
        label.pack(side="left")
        rm_btn = tk.Button(line_frame, text="X", bg="red", fg="white")
        rm_btn.pack(side="right")
        rm_btn.bind("<Button-1>", lambda event, index=index: remove_filepath(event, index))
        
def confirm_selection():
        print("selected list of filepath(s) {}".format(selected_filepaths))
        root.destroy()
    
def remove_filepath(event, index):
    #handling_widget = event.widget
    #line_frame = handling_widget.master
    #line_frame.destroy()
    print("removed {}".format(selected_filepaths[index]))
    del selected_filepaths[index]
    clear_list()
    build_list()

def open_browser():
    clear_list()
        
    filepath = filedialog.askopenfilename(initialdir="/", title="Select File",
                                         filetypes=(accepted_file_formats))
    if (filepath != ""):
        selected_filepaths.append(filepath)
        print("selected {}".format(filepath))
    else:
        print("no file selected")
    
    build_list()
        
canvas = tk.Canvas(root, bg="#ADD8E6")
canvas.place(relwidth=1, relheight=1, relx=0.5, rely=0.5, anchor="center")

frame = tk.Frame(root, bg="white")
frame.place(relwidth=0.8, relheight=0.8, relx=0.1, rely=0.1)

# Scrollbar only works on canvas
# Thus, we create a list_frame on a list_canvas on a frame
list_canvas = tk.Canvas(frame)
list_frame = tk.Frame(list_canvas)
scrollbar = tk.Scrollbar(frame, orient="vertical", command=list_canvas.yview)

list_canvas.configure(yscrollcommand=scrollbar.set)
scrollbar.pack(side="right", fill="y")
list_canvas.pack(side="left", fill="both", expand=True)
list_canvas.create_window((0,0), window=list_frame, anchor='nw')

list_frame.bind("<Configure>", lambda event, canvas=list_canvas: onFrameConfigure(canvas))

# Buttons
browse_btn = tk.Button(root, text="Browse", padx=10, pady=5, 
                     fg="white", bg="black",
                    command=open_browser)
browse_btn.place(relx=0.40, rely=1, anchor="s")

confirm_btn = tk.Button(root, text="Confirm", padx=10, pady=5,
                              fg="white", bg="black",
                              command=confirm_selection)
confirm_btn.place(relx=0.60, rely=1, anchor="s")

# Start
root.mainloop()

print(selected_filepaths)
```
